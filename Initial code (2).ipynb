```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
gamma = .1
mu = 1
diameter = .0000001
mp = 1.67*10**-27
beta = 3*np.pi*mu*diameter/mp
kb = 1.38*10**-23
T = 295.372222
cc = 11
S = 2*kb*T*beta/mp*cc
deltaT = np.linspace(0.001, 2*10**-6, num=1000)
Fb = gamma*((np.pi*S)/(deltaT))**.5

plt.plot(deltaT, Fb)
#plt.show()
```

```python
x1 = 0
u1 = 0
U_ = []
X_ = []
Fb = np.array(Fb)
deltaT = np.array(deltaT)
for j in deltaT:
    U = []
    X = []
    U.append(u1)
    X.append(x1)
    for i in range(0,1000):
        u2 = U[i] + j*Fb[i]
        x2 = X[i] + u2*j
        X.append(x2)
        U.append(u2)
    X_.append(X)
    U_.append(U)
plt.plot(np.mean(X_, axis=1), np.mean(U_, axis=1))
```
